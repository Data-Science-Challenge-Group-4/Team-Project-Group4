# **Import Libraries**


```python
#Importing libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score, RandomizedSearchCV
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.metrics import (
    accuracy_score, f1_score, roc_auc_score, classification_report,
    confusion_matrix, ConfusionMatrixDisplay, roc_curve
)
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from xgboost import XGBClassifier
from lightgbm import LGBMClassifier

import warnings
warnings.filterwarnings("ignore")
sns.set(style="whitegrid", palette="pastel")
```

# **Load and Inspect Data**

```python
#  LOAD DATA

train_df = pd.read_csv("dataset_B_training.csv")
test_df = pd.read_csv("dataset_B_testing.csv")
example_sub = pd.read_csv("dataset_B_example_submission.csv")

print(" Training data:", train_df.shape)
print(" Test data:", test_df.shape)
train_df.head()
```

# **Exploratory Data Analysis (EDA)**

```python
#  Basic info
train_df.info()
train_df.describe(include="all").T.head(15)

#  Target distribution
plt.figure(figsize=(5,4))
sns.countplot(x="h1n1_vaccine", data=train_df, palette="coolwarm")
plt.title("Target Distribution â€” H1N1 Vaccine")
plt.show()

#  Missing Values
missing = train_df.isnull().mean().sort_values(ascending=False)
plt.figure(figsize=(8,4))
missing.head(20).plot(kind="bar", color="tomato")
plt.title("Top Missing Values")
plt.ylabel("Proportion Missing")
plt.show()

#  Identify numeric & categorical columns
cat_cols = train_df.select_dtypes(include="object").columns.tolist()
num_cols = train_df.select_dtypes(include=["int64","float64"]).columns.drop("h1n1_vaccine").tolist()
print(f"Categorical features: {len(cat_cols)}")
print(f"Numerical features: {len(num_cols)}")

# Numerical distributions
plt.rcParams.update({
    'axes.titlesize': 10,
    'axes.labelsize': 8,
    'xtick.labelsize': 7,
    'ytick.labelsize': 7,
    'legend.fontsize': 8
})

train_df[num_cols].hist(bins=25, figsize=(12,10), color="skyblue", edgecolor="black")
plt.suptitle("Numerical Feature Distributions", fontsize=12)
plt.tight_layout(pad=2)
plt.show()

#  Correlation heatmap
plt.figure(figsize=(10,8))
sns.heatmap(train_df[num_cols + ["h1n1_vaccine"]].corr(), cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

# **Data Preprocessing**

```python
TARGET_COL = "h1n1_vaccine"
ID_COL = "respondent_id"

X = train_df.drop(columns=[TARGET_COL, ID_COL])
y = train_df[TARGET_COL]

# Identify numeric & categorical columns after dropping target and ID
cat_cols = X.select_dtypes(include="object").columns.tolist()
num_cols = X.select_dtypes(include=["int64","float64"]).columns.tolist()

num_pipe = Pipeline([
    ("imp", SimpleImputer(strategy="median")),
    ("scaler", StandardScaler())
])

cat_pipe = Pipeline([
    ("imp", SimpleImputer(strategy="most_frequent")),
    ("ohe", OneHotEncoder(handle_unknown="ignore", sparse_output=False))
])

base_preprocessor = ColumnTransformer([
    ("num", num_pipe, num_cols),
    ("cat", cat_pipe, cat_cols)
])

X_train, X_valid, y_train, y_valid = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
print(" Train:", X_train.shape, " | Valid:", X_valid.shape)
```

# **Model Building**

```python
# Model Building

def evaluate_model(model, X_train, y_train, X_valid, y_valid, name="Model"):
    model.fit(X_train, y_train)
    y_pred = model.predict(X_valid)
    y_prob = model.predict_proba(X_valid)[:,1] if hasattr(model, "predict_proba") else None
    acc = accuracy_score(y_valid, y_pred)
    f1  = f1_score(y_valid, y_pred)
    auc = roc_auc_score(y_valid, y_prob) if y_prob is not None else np.nan
    print(f"{name}: Acc {acc:.3f} | F1 {f1:.3f} | AUC {auc:.3f}")
    return {"Model": name, "Accuracy": acc, "F1": f1, "AUC": auc}

models = {
    "Logistic Regression": Pipeline([("prep", base_preprocessor),
        ("clf", LogisticRegression(max_iter=1000, class_weight="balanced"))]),
    "Random Forest": Pipeline([("prep", base_preprocessor),
        ("clf", RandomForestClassifier(n_estimators=400, random_state=42, class_weight="balanced"))]),
    "Gradient Boosting": Pipeline([("prep", base_preprocessor),
        ("clf", GradientBoostingClassifier(random_state=42))]),
    "XGBoost": Pipeline([("prep", base_preprocessor),
        ("clf", XGBClassifier(random_state=42, n_estimators=500, learning_rate=0.05,
                               max_depth=5, subsample=0.9, colsample_bytree=0.9,
                               eval_metric="logloss"))]),
    "LightGBM": Pipeline([("prep", base_preprocessor),
        ("clf", LGBMClassifier(random_state=42, n_estimators=500,
                               learning_rate=0.05, subsample=0.9, colsample_bytree=0.9))])
}

results = []
for name, model in models.items():
    results.append(evaluate_model(model, X_train, y_train, X_valid, y_valid, name))

results_df = pd.DataFrame(results).sort_values("AUC", ascending=False)
display(results_df)

plt.figure(figsize=(7,4))
sns.barplot(x="AUC", y="Model", data=results_df, palette="Blues_r")
plt.title("Model Comparison (AUC)")
plt.show()
```

# **Performance Evaluation and Tuning**

```python
#Performance Evaluation and hyperparameter tuning

best_model_name = results_df.iloc[0]["Model"]
best_model = models[best_model_name]
best_model.fit(X_train, y_train)

y_pred = best_model.predict(X_valid)
y_prob = best_model.predict_proba(X_valid)[:,1]
print(f" Best: {best_model_name} | AUC={roc_auc_score(y_valid,y_prob):.3f}")

# Confusion Matrix
cm = confusion_matrix(y_valid, y_pred)
ConfusionMatrixDisplay(cm, display_labels=["No Vaccine","Got Vaccine"]).plot(cmap="coolwarm")
plt.title(f"{best_model_name} Confusion Matrix")
plt.show()

# ROC Curve
fpr, tpr, _ = roc_curve(y_valid, y_prob)
plt.plot(fpr, tpr, label=f"{best_model_name} (AUC={roc_auc_score(y_valid,y_prob):.3f})")
plt.plot([0,1],[0,1],'--',color='grey')
plt.legend(); plt.title("ROC Curve â€“ Validation"); plt.show()

# Cross-validation
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
cv_auc = cross_val_score(best_model, X, y, scoring="roc_auc", cv=cv)
print(f"5-fold AUC: {cv_auc.mean():.4f} Â± {cv_auc.std():.4f}")

# hyper-parameter tuning
rf = Pipeline([("prep", base_preprocessor),
    ("clf", RandomForestClassifier(random_state=42,class_weight="balanced"))])
param = {"clf__n_estimators":[200,400,600],"clf__max_depth":[5,10,None],"clf__min_samples_split":[2,5,10]}
search = RandomizedSearchCV(rf,param_distributions=param,n_iter=6,scoring="roc_auc",cv=3,n_jobs=-1,random_state=42)
search.fit(X_train,y_train)
print("Best RF params:",search.best_params_,"| AUC:",search.best_score_)
```

# **Final Submission**

```python
#Final Submission of files
ID_COL = "respondent_id"
X_test = test_df.drop(columns=[ID_COL])
test_ids = test_df[ID_COL]

top5 = results_df["Model"].head(5).tolist()
submissions = {}

for i, name in enumerate(top5, start=1):
    model = models[name]
    model.fit(X, y)
    preds = model.predict_proba(X_test)[:,1]
    sub = pd.DataFrame({ID_COL: test_ids, "h1n1_vaccine": preds})
    fname = f"submission_{i}_{name.replace(' ','_').lower()}.csv"
    sub.to_csv(fname, index=False)
    submissions[name] = sub
    print(f" Saved: {fname}")

summary = results_df.set_index("Model").loc[top5].reset_index()
summary["Submission_File"] = [f"submission_{i}_{n.replace(' ','_').lower()}.csv" for i,n in enumerate(top5, 1)]
summary.to_csv("submission_summary.csv", index=False)
display(summary)

plt.figure(figsize=(10,6))
for n in top5:
    sns.kdeplot(submissions[n]["h1n1_vaccine"], label=n)
plt.title("Predicted Probability Distributions â€“ Top Models")
plt.legend(); plt.show()

print("\n All 5 submission files and summary CSV ready for Canvas upload.")
```

## Helper Functions for Pipelines and Feature Names
We create small utility functions to extract the model components (preprocessor and classifier) from our sklearn pipeline.  
These functions also retrieve the feature names after One-Hot Encoding, which will be used later for interpretability.


```python
# === HELPER FUNCTIONS ===

def get_pipeline_parts(pipeline):
    """Extract preprocessor and classifier from Pipeline."""
    prep = pipeline.named_steps.get("prep", None) or pipeline.named_steps.get("preprocessor", None)
    clf = pipeline.named_steps.get("clf", pipeline.steps[-1][1])
    return prep, clf

def get_feature_names_from_ct(ct: ColumnTransformer, input_features: list):
    """Get output feature names after ColumnTransformer."""
    names = []
    for name, transformer, cols in ct.transformers_:
        if transformer == "drop":
            continue
        if transformer == "passthrough":
            names.extend(list(cols))
            continue
        final = transformer
        if hasattr(transformer, "steps"):
            final = transformer.steps[-1][1]
        if isinstance(final, OneHotEncoder):
            ohe_cols = final.get_feature_names_out(cols)
            names.extend(ohe_cols.tolist())
        else:
            names.extend(list(cols if isinstance(cols, (list, tuple)) else [cols]))
    return names
```

## Preparing the Best Model and Data for Explainability
In this step, we extract our best-performing model and prepare the training and validation data for SHAP analysis.  
We ensure the preprocessor is fitted and transform both datasets to numeric arrays for SHAP interpretation.


```python
# === PREPARE BEST MODEL AND DATA ===

prep, clf = get_pipeline_parts(best_model)

# Fit preprocessor if needed
try:
    _ = prep.transform(X_train.head(1))
except Exception:
    prep.fit(X_train, y_train)

X_train_t = prep.transform(X_train)
X_valid_t = prep.transform(X_valid)

try:
    feature_names = get_feature_names_from_ct(prep, X_train.columns.tolist())
except Exception:
    feature_names = [f"f{i}" for i in range(X_train_t.shape[1])]

print("Data ready for explainability.")
```

## Setting up the SHAP Explainer
We initialise the appropriate SHAP explainer based on the model type (tree-based, linear, or other).  
This allows us to compute feature-level attributions that explain how each input influences model predictions.


```python
# === SHAP EXPLAINER SETUP ===

model_name_lower = clf.__class__.__name__.lower()
is_tree = any(k in model_name_lower for k in ["forest", "xgb", "lgbm", "gradientboost", "decisiontree"])

def get_pos_class_shap(sv, explainer):
    """Return SHAP values for positive class if a list is returned."""
    if isinstance(sv, list):
        if len(sv) == 2:
            return sv[1], np.atleast_1d(explainer.expected_value)[1] if np.ndim(explainer.expected_value) > 0 else explainer.expected_value
        return sv[-1], np.atleast_1d(explainer.expected_value)[-1] if np.ndim(explainer.expected_value) > 0 else explainer.expected_value
    return sv, explainer.expected_value

if is_tree:
    explainer = shap.TreeExplainer(clf)
    shap_values = explainer.shap_values(X_valid_t)
    sv_pos, expected_pos = get_pos_class_shap(shap_values, explainer)
else:
    from sklearn.linear_model import LogisticRegression
    if isinstance(clf, LogisticRegression):
        bg = shap.sample(X_train_t, 500, random_state=42)
        explainer = shap.LinearExplainer(clf, bg)
        sv_pos = explainer.shap_values(X_valid_t)
        expected_pos = explainer.expected_value
    else:
        bg = shap.sample(X_train_t, 200, random_state=42)
        explainer = shap.KernelExplainer(lambda z: clf.predict_proba(z)[:,1], bg)
        sv_pos = explainer.shap_values(X_valid_t[:200])
        X_valid_t = X_valid_t[:200]
        expected_pos = explainer.expected_value

print("SHAP explainer ready.")
```

## Global Feature Importance using SHAP
We generate SHAP summary and bar plots to show the most influential features across the entire validation dataset.  
These global explanations highlight which variables most strongly affect the modelâ€™s prediction of H1N1 vaccination uptake.


```python
# === GLOBAL SHAP PLOTS ===

plt.figure(figsize=(9, 6))
shap.summary_plot(sv_pos, features=X_valid_t, feature_names=feature_names, show=False, max_display=25)
plt.title("SHAP Summary â€” Global Feature Impact")
savefig("fig_shap_summary")
plt.show()

plt.figure(figsize=(9, 6))
shap.summary_plot(sv_pos, features=X_valid_t, feature_names=feature_names, plot_type="bar", show=False, max_display=25)
plt.title("SHAP Bar Plot â€” Mean |SHAP| Importance")
savefig("fig_shap_bar")
plt.show()
```

## SHAP Dependence Plots for Key Features
In this section, we visualise the relationship between individual features and their SHAP values.  
These plots show how increasing or decreasing feature values impacts the modelâ€™s output probability.


```python
# === SHAP DEPENDENCE PLOTS (TOP 2 FEATURES) ===

mean_abs = np.abs(sv_pos).mean(axis=0)
top_idx = np.argsort(-mean_abs)[:2]
top_features = [feature_names[i] for i in top_idx]

for f in top_features:
    plt.figure(figsize=(7, 5))
    shap.dependence_plot(f, sv_pos, X_valid_t, feature_names=feature_names, show=False)
    plt.title(f"SHAP Dependence â€” {f}")
    savefig(f"fig_shap_dependence_{f.replace('/', '_')}")
    plt.show()
```

## Local Explanations with SHAP Waterfall Plots
Finally, we generate local explanations for a few individual samples.  
Each waterfall plot shows how specific features contribute positively or negatively to a personâ€™s predicted probability of vaccination.


```python
# === LOCAL SHAP WATERFALLS (2 RANDOM SAMPLES) ===

idxs = np.random.RandomState(42).choice(len(X_valid_t), size=min(2, len(X_valid_t)), replace=False)

for i in idxs:
    base_val = np.atleast_1d(expected_pos)[0] if np.ndim(expected_pos) > 0 else expected_pos
    exp = shap.Explanation(
        values=sv_pos[i],
        base_values=base_val,
        data=X_valid_t[i],
        feature_names=feature_names
    )
    shap.plots.waterfall(exp, max_display=12, show=False)
    plt.title(f"Local SHAP Waterfall â€” Sample {i}")
    savefig(f"fig_shap_local_{i}")
    plt.show()

print("\nAll SHAP plots complete and saved as PNGs.")
```
